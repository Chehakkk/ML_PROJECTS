Beginning with robust data preprocessing—including exploratory analysis, irrelevant column removal, and feature scaling via StandardScaler—the workflow guides users through KMeans clustering supported by the Elbow Method to determine optimal cluster count.
```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python

from sklearn.cluster import KMeans
```

```python
df=pd.read_excel("University_Clustering.xlsx")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop("State",axis=1,inplace=True)
```

```python
new_df=df.drop("Univ",axis=1)
new_df
```

```python
from sklearn.preprocessing  import StandardScaler
```

```python
for col in new_df:
    ss= StandardScaler()
    new_df[col]=ss.fit_transform(new_df[[col]])
```

```python
wcss=[]
clusters=list(range(1,9))
for K in clusters:
    model=KMeans(n_clusters=K)
    model.fit(new_df)
    wcss.append(model.inertia_)
```

```python
plt.figure()
plt.plot(clusters,wcss,"x-")
plt.xlabel("K")
plt.ylabel("WCSS")
plt.show()
```

albo method is graphical representation of finding the optimal K in k means clustering it works by finding WCSS  that is the sum of square distance between points in a cluster and the clusters centroid 

```python
model=KMeans(n_clusters=3)
model.fit(new_df)
```

```python
model.labels_
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3,random_state=1)
```

```python
components=pca.fit_transform(new_df)
components
```

```python
pc1=components[:,0]
pc2=components[:,1]
pc3=components[:,2]
```

```python
fig=plt.figure(figsize=(8,8))
ax=plt.axes(projection='3d')
ax.scatter(pc1,pc2,pc3 ,c=df['labels'])
plt.show()
```
